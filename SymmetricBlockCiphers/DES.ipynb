```python
# PT = input("Enter plain text (64 bits): ")
PT = "0000000100100011010001010110011110001001101010111100110111101111"
# key = input("Enter Master Key (64 bits): ")
key = "0001001100110100010101110111100110011011101111001101111111110001"
```

```python
#subkey generation

sub_keys = []
sub_key = "".join(key[i-1] for i in PC1.ravel())
for i in range(16):
    L = sub_key[:28]
    R = sub_key[28:]
    
    #left circular shift
    bits = LCS[i]
    L = "".join(L[bits:]+L[:bits])
    R = "".join(R[bits:]+R[:bits])
    sub_key = "".join(L+R)
    
    #permuting of left shifted subkey
    temp = "".join([sub_key[i-1] for i in PC2.ravel()])
    sub_keys += [temp]
```

```python
#initial permutation

PT = "".join(PT[i-1] for i in IP.ravel())
```

```python
def XOR(b1,b2, l):
    result = ""
    for i in range(l):
        if(b1[i]==b2[i]):
            result += '0'
        else:
            result += '1'
    return result
```

```python
#round function

L = PT[:32]
R = PT[32:]
for i in range(16):
    R_temp = "".join(R[i-1] for i in EXP.ravel()) #expansion
    R_temp = XOR(R_temp, sub_keys[i], 48) #XOR with sub key
    R_sub = [
        R_temp[:6],
        R_temp[6:12],
        R_temp[12:18],
        R_temp[18:24],
        R_temp[24:30],
        R_temp[30:36],
        R_temp[36:42],
        R_temp[42:]
    ]
    for j in range(8): #substitution
        sub = R_sub[j]
        row = int(sub[0]+sub[5], 2)
        col = int(sub[1:5], 2)
        R_sub[j] = bin(S[j][row][col])[2:]
        R_sub[j] = "0"*(4-len(R_sub[j]))+R_sub[j]
    R_sub = "".join(R_sub)
    R_temp = "".join(R_sub[i-1] for i in P.ravel()) #permutation
    R_temp = XOR(L, R_temp, 32) #XOR with L side
    
    
    L = R
    R = R_temp
    
new_PT = R+L
```

```python
# inverse permutation

cipher = "".join([new_PT[i-1] for i in InvIP.ravel()])
print(f"Cipher is {cipher}")
```
